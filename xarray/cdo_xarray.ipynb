#### Subsetting data

```python
pr = prEnsMean.pr
```

```python
pr.shape # numpy operation
```

```python
pr[0,:] # at a specific time stamp
```

```python
pr[0,:].time
```

```python
from datetime import datetime
```

##### Slicing per 30 years
{2011 - 2040} 
\\
{2041 - 2070}
\\
{2071 - 2100}

```python
batch1140 = pr.sel(time=slice(datetime(2011,1,1),datetime(2041,1,1)))
print('\n STARTING DATE \n: ',batch1140[0].time, '\n END DATE: \n',batch1140[-1].time)
```

```python
batch4170 = pr.sel(time=slice(datetime(2041,1,1),datetime(2071,1,1)))
print('\n STARTING DATE \n: ',batch4170[0].time, '\n END DATE: \n',batch4170[-1].time)
```

```python
batch7100 = pr.sel(time=slice(datetime(2071,1,1),datetime(2101,1,1)))
print('\n STARTING DATE \n: ',batch7100[0].time, '\n END DATE: \n',batch7100[-1].time)
```

#### Plotting

```python
batch1140.mean(dim='time').plot(size=3)
```

```python
batch4170.mean(dim='time').plot(size=3)
```

```python
batch7100.mean(dim='time').plot(size=3)
```

```python
batch1140.mean(dim=('rlat','rlon')).plot(size=5)
```

```python
batch4170.mean(dim=('rlat','rlon')).plot(size=5)
```

```python
batch7100.mean(dim=('rlat','rlon')).plot(size=5)
```

#### Grouping and resampling

##### Calculating anomalies
1. Between 2011-40 & 2041-2070 
2. Between 2041-2070 & 2071-2100

```python
(pr - batch4170).mean(dim=('rlat', 'rlon')).plot(size=4, label='Anomaly')
```

```python
(pr - batch7100).mean(dim=('rlat', 'rlon')).plot(size=4)
```

##### Calculate seasonal mean

```python
batch1140.groupby('time.season').mean(dim='time').plot(col='season', col_wrap=4)
```

```python
batch4170.groupby('time.season').mean(dim='time').plot(col='season', col_wrap=4)
```

```python
batch7100.groupby('time.season').mean(dim='time').plot(col='season', col_wrap=4)
```

#### Masking

```python
from numpy import testing
testing.assert_array_equal(batch4170, batch7100)
```

```python
pr.groupby('time.season').mean(dim='time').plot(col='season', col_wrap=4)
```

#### Writing to files

```python
batch1140.to_netcdf('batch1140_xarray.nc')
batch4170.to_netcdf('batch4170_xarray.nc')
batch7100.to_netcdf('batch7100_xarray.nc')
```

#### Reading from multiple files

```python
files = [
    'batch1140_xarray.nc',
    'batch4170_xarray.nc',
    'batch7100_xarray.nc'
]
```

```python
prFromBatches = xarray.open_mfdataset(files)
```

```python
print('STARTING DATE \n', prFromBatches.pr[0].time , '\n END DATE \n', prFromBatches.pr[-1].time)
```

```python
prFromBatches.pr.groupby('time.season').mean(dim='time').plot(col='season', col_wrap=4)
```

```python
pr.sel(time=slice(datetime(2011,1,1),datetime(2101,1,1))).identical(prFromBatches.pr)
```
